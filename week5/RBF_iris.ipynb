```python
import numpy as np
import numpy.random as ra
import sklearn.datasets as da
import pandas as pd
from sklearn.cluster import KMeans
```

```python
def rbf_train(input_vector,output_vector,lr,hnum,iter_num):
    weights1 = np.array(np.random.rand(len(input_vector[0]),hnum))
    weights2 = np.array(np.random.rand(hnum+1,len(output_vector[0])))
    
    w=KMeans(n_clusters=hnum).fit(input_vector)
    weights1= np.transpose(w.cluster_centers_)    
    
    print(weights1)
    l=np.zeros((len(input_vector),hnum+1))
    d = (input_vector.max(axis=0)-input_vector.min(axis=0)).max()
    sigma = d/np.sqrt(2*hnum)  
    for i in range(hnum):
        l[:,i] = np.exp(-np.sum((input_vector-np.ones((1,len(input_vector[0])))*weights1[:,i])**2,axis=1)/(2*sigma**2))
    l[:,:-1] /= np.transpose(np.ones((1,len(l)))*l[:,:-1].sum(axis=1))
    print(l)
    l=np.insert(l,0,-1,axis=1)
    weights2=np.dot(np.linalg.pinv(l),output_vector)
    print(weights2)
    return weights1,weights2,sigma
        
```

```python
def predict(inp,w1,w2):
    hidact=np.zeros((len(inp),hnum+1))
    
    
    for i in range(hnum):
        hidact[:,i] = np.exp(-np.sum((inp - np.ones((1,len(inp[0])))*w1[:,i])**2,axis=1)/(2*sigma**2))
    
    hidact=np.insert(hidact,0,-1,axis=1)
    hidact[:,:-1] /= np.transpose(np.ones((1,len(hidact)))*hidact[:,:-1].sum(axis=1))
    out=np.dot(hidact,w2)
    
    output = np.where(out>0,1,0)
    
    return output 
    
    
```

```python
d = da.load_iris()
input_vector = d['data'][:,:4]
output_vector = np.array(pd.get_dummies(d['target']))
print(input_vector)
print("\n")
print(output_vector)
```

```python
v = np.concatenate((input_vector,output_vector),axis=1)

ra.shuffle(v)

train_in = v[:101,:4]
train_out = v[:101,4:]
test_in = v[101:,:4]
test_out = v[101:,4:]
```

```python
hnum=15
lr = 0.01
iter_num = 10
w1,w2,sigma=rbf_train(train_in,train_out,lr,hnum,iter_num)
print(sigma)
```

```python
y=predict(test_in,w1,w2)
a=[]
for i in range(len(y)):
    a.append(((y[i]==test_out[i]).sum()==3))
print(np.array(a).sum()/len(y))
```
