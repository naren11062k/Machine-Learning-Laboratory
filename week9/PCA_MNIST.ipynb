```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('train.csv')
print(data.head(5))
l = data['label']
d = data.drop('label',axis=1 )
```

```python
lab = l.head(15000)
dat = d.head(15000)
```

```python
std_data = StandardScaler().fit_transform(dat)
std_data.shape
```

```python
pca = PCA()
```

```python
pca.n_components = 2
pca_data = pca.fit_transform(std_data)
```

```python
pca_data = np.vstack((pca_data.T,lab)).T
```

```python
df = pd.DataFrame(data= pca_data, columns= ('1st principle','2nd principle', 'label'))
sn.FacetGrid(df,hue='label', height=6).map(plt.scatter,'1st principle','2nd principle').add_legend
plt.show()
```
